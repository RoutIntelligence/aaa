# [dbo].[DevoluçãoProduto]

```sql
/****** Object:  StoredProcedure [dbo].[DevoluçãoProduto]    Script Date: 19/11/2025 09:51:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[DevoluçãoProduto]
@ProdutoID int,
@PedidoID  int

as
BEGIN 
      Declare  @Qnt int

	  select Quantidade = @QNT
	  from ItemPedido
	  where ProdutoID = @produtoID and PedidoID = @pedidoid

	  Delete from ItemPedido
	  where ProdutoID = @produtoID and PedidoID = @pedidoid

	  update Produto
	  set Estoque = Estoque + @qnt
	  where Produtoid = @produtoid
END
GO
```

# [dbo].[HistóricoCliente]

```sql
/****** Object:  StoredProcedure [dbo].[HistóricoCliente]    Script Date: 19/11/2025 09:51:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[HistóricoCliente]
@ClienteId int

as
BEGIN  
      SELECT c.Nome,
           p.PedidoID,
           p.DataPedido,
           SUM(ip.Quantidade * pr.Preco) AS ValorPedido
    FROM Cliente c
    JOIN Pedido p ON c.ClienteID = p.ClienteID
    JOIN ItemPedido ip ON p.PedidoID = ip.PedidoID
    JOIN Produto pr ON ip.ProdutoID = pr.ProdutoID
    WHERE c.ClienteID = @ClienteID
    GROUP BY c.Nome, p.PedidoID, p.DataPedido;

    -- Total geral
    SELECT c.Nome,
           SUM(ip.Quantidade * pr.Preco) AS TotalGasto
    FROM Cliente c
    JOIN Pedido p ON c.ClienteID = p.ClienteID
    JOIN ItemPedido ip ON p.PedidoID = ip.PedidoID
    JOIN Produto pr ON ip.ProdutoID = pr.ProdutoID
    WHERE c.ClienteID = @ClienteID
    GROUP BY c.Nome;
END;
GO
```

# [dbo].[sp_TopProdutos]

```sql
/****** Object:  StoredProcedure [dbo].[sp_TopProdutos]    Script Date: 19/11/2025 09:51:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   PROCEDURE [dbo].[sp_TopProdutos]
    @TopN INT
AS
BEGIN
    SELECT TOP (@TopN) pr.Nome,
           SUM(ip.Quantidade) AS TotalVendido
    FROM Produto pr
    JOIN ItemPedido ip ON pr.ProdutoID = ip.ProdutoID
    GROUP BY pr.Nome
    ORDER BY TotalVendido DESC;
END;
GO
```

# [dbo].[VendasCategoria]

```sql
/****** Object:  StoredProcedure [dbo].[VendasCategoria]    Script Date: 19/11/2025 09:51:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[VendasCategoria]
@Categoria nvarchar(50)
as
BEGIN
     select pr.Nome, sum(quantidade) as 'quantidadeTotal', sum(quantidade * preço) as 'ValorTotal'
	 from ItemPedido it join Produto pr on it.ProdutoID = pr.ProdutoID
	 where Categoria = @categoria
	 group by pr.Nome
END
GO
```
