```python
------------------------------------------------
#Analise da Categoria mais vendida

counts_Category=sp['Category'].value_counts()
```

```python
plt.pie(counts_Category,labels=sp['Category'].value_counts().index,
        autopct='%1.2f%%')

plt.title('Categoria Mais Vendida')
plt.show()
```

```python
-----------------------------------------------------------------------------
#Soma das Vendas por categoria

order = sp.groupby('Category')['Sales'].sum().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=sp,
    x='Category',
    y='Sales',
    estimator=np.sum,
    order=order,
    hue='Category',
    palette='cool',
    legend=False
)
plt.title('Soma de Vendas')
plt.xlabel('Categoria')
plt.ylabel('Vendas')
----------------------------------------
# Technology maior número de vendas
```

```python
-----------------------------------------------------------------------------
#Soma de Lucro por Categoria

Lucro_Categoria=sp.groupby('Category')['Profit'].sum().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=sp,
    x='Category',
    y='Profit',
    estimator=np.sum,
    order=Lucro_Categoria,
    hue='Category',
    palette='cool',
    
)
plt.title('Lucro por Categoria')
plt.show()
--------------------------------------------
# Maior Categoria com Ganhos é 'Technology'
# Menor Categoria com Ganhos é 'Furniture'
```

```python
# Vendas por Sub-Category

plt.figure(figsize=(8,5))
sns.countplot(
    data=sp,
    x='Sub-Category',
    order=sp['Sub-Category'].value_counts().index,
    hue='Sub-Category',
    palette='cool'
    
)
plt.xticks(rotation=90)
plt.title('Vendas por Sub-Category')
plt.show()
------------------------------------------
# Maior Venda como Sub-Category é 'Binders'
# Menor Venda como Sub-Category é 'Paper'
```

```python
# Vendas totais por Sub-Category
Vendas_SubCategoria=sp.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=sp,
    x='Sub-Category',
    y='Sales',
    estimator=np.sum,
    order=Vendas_SubCategoria,
    hue='Sub-Category',
    palette='cool',
)
plt.title('Soma de Vendas Sub-Category')
plt.xticks(rotation=90)
plt.show()
---------------------------------------------------------
# Maior número de vendas está entre 'Phones' e 'Chairs'
# Menor número de vendas é 'Fasteners'
```

```python
profit_SubCategory=sp.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=sp,
    x='Sub-Category',
    y='Profit',
    estimator=np.sum,
    order=profit_SubCategory,
    hue='Sub-Category',
    palette='cool',
    legend=False
)
plt.title('Soma de Lucro de Sub-Category')
plt.xticks(rotation=90)
plt.show()
----------------------------------------------------
# Maior lucro de Sub-Category é 'Copiers'
# Menor lucro de Sub-Category 'Machines' e 'Fasteners' (lucro é quase nulo)
# Sub-Category que geram prejuizos são 'Supplies', 'Bookcases' e 'Tables'
```

```python
--------------------------------------------------------
#Produtos Mais Vendidos de 2017 entre o mês 10 e 11

sp['Order Date']=pd.to_datetime(sp['Order Date'],errors='coerce')
```

```python
Ano2017_10_11=sp[(sp['Order Date'].dt.year== 2017) &
(sp['Order Date'].dt.month.isin([10,11]))
].copy()
```

```python
Ano2017_10_11['Month5']=Ano2017_10_11['Order Date'].dt.month
```

```python
vendas_Mês10e11_Ano2017=Ano2017_10_11.groupby(['Month5','Product Name'])['Sales'].sum().reset_index()
```

```python
top5_produtos=vendas_Mês10e11_Ano2017.groupby('Product Name')['Sales'].sum().nlargest(5).index
```

```python
vendas_top5_2017_Mês10e11=vendas_Mês10e11_Ano2017[vendas_Mês10e11_Ano2017['Product Name'].isin(top5_produtos)]
```
